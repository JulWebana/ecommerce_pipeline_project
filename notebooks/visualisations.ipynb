Importons les bibliothèques nécessaires

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style plus joli
sns.set(style="whitegrid")
```

### Top produits par chiffre d'affaires

```python
df = pd.read_csv("../output/top_products.csv")

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x="total_revenue",
    y="product",
    hue="product",
    palette="viridis",
    legend=False
)
plt.title("Top 5 produits par chiffre d’affaires")
plt.xlabel("Chiffre d'affaires (€)")
plt.ylabel("Produit")
plt.tight_layout()
plt.show()
```

Les 5 produits ayant généré le plus de revenus (pas nécessairement les plus vendus).

##### Interprétation :

- Les headphones sont le produit le plus rentable.

- Les produits : keyboard, Phone et Mouse suivent de près, avec des chiffres d’affaires assez proches.

- Le produit Laptop est 5e malgré son prix élevé:  probablement moins vendu que les autres.

##### Conclusion business :

- Il est intéressant de creuser pourquoi les Headphones performent aussi bien (offre, popularité, marges ?).
- Les produits entre la 2e et 4e place sont très proches : une promo peut inverser le classement.



### Ventes par ville

```python
df = pd.read_csv("../output/sales_by_city.csv")

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x="total_revenue",
    y="city",
    hue="city",            
    palette="crest",
    legend=False            
)
plt.title("Ventes par ville")
plt.xlabel("Chiffre d'affaires (€)")
plt.ylabel("Ville")
plt.tight_layout()
plt.show()
```

##### Interprétation :

- Nantes est la ville n°1 en terme de chiffre d’affaires 

- Lyon, Bordeaux, Rennes et Lille forment un peloton solide.

- Nice, Paris et Unknown ferment la marche.

##### Conclusion business :

- Nantes semble être un marché stratégique. Il faudrait analyser les profils clients de cette ville.

- La part importante de "Unknown" peut signaler des données manquantes : à traiter ou enrichir.

### Évolution des ventes quotidiennes

```python
df = pd.read_csv("../output/daily_sales.csv")
df["order_date"] = pd.to_datetime(df["order_date"])

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="order_date", y="daily_revenue", marker="o")
plt.title("Évolution des ventes quotidiennes")
plt.xlabel("Date")
plt.ylabel("Chiffre d'affaires (€)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Le graphique nous renseigne sur le chiffre d’affaires par jour, sur une période de 2 mois environ.

### Interprétation :

- Le CA est très fluctuant, avec des pics > 20 000 € et des jours à 0.

- Il y a des hausses soudaines, peut-être liées à des promos ou campagnes.

- Le mois de février semble un peu plus actif que mars.


### Conclusion business :

- Le modèle de vente est sans doute irrégulier (type flash sales, ou site e-commerce non linéaire c-à-d pics soudains de ventes, ventes liées à des événements).

- On pourrait calculer la moyenne mobile pour lisser les tendances.



### Panier moyen par client

```python
df = pd.read_csv("../output/avg_basket_per_customer.csv")

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x="avg_basket",
    y="customer_name",
    hue="customer_name",   
    palette="mako",
    legend=False           
)
plt.title("Panier moyen par client (Top 10)")
plt.xlabel("Panier moyen (€)")
plt.ylabel("Client")
plt.tight_layout()
plt.show()
```

Le graphique montre le montant moyen dépensé par commande pour chaque client du top 10.


##### Interprétation :

- Nathan est le client le plus rentable par commande (~2800€)

- Tous les clients du top 10 ont un panier moyen supérieur à 2000€, ce qui est très élevé.


##### Conclusion business :

- Ces clients sont peut-être des entreprises ou des clients fidèles B2B.

- On pourrait cibler ces clients avec un programme VIP ou des offres personnalisées.

### Produits les plus commandés (par quantité)

```python
df = pd.read_csv("../output/top_products_by_quantity.csv")

plt.figure(figsize=(8, 5))
sns.barplot(
    data=df,
    x="total_quantity",
    y="product",
    hue="product",       
    palette="rocket",
    legend=False         
)
plt.title("Produits les plus commandés (par quantité)")
plt.xlabel("Quantité totale vendue")
plt.ylabel("Produit")
plt.tight_layout()
plt.show()
```

Le graphique nous renseigne sur le nombre d’unités vendues par produit, sans tenir compte du prix.


##### Interprétation :

- Les Headphones sont les plus vendus à la fois en quantité et en CA (star du catalogue).

- Les produits comme Mouse, Keyboard, Phone suivent, ce qui est cohérent avec leur place dans le CA.

- Des produits comme le Charger et la Smartwatch ferment la marche : soit moins populaires, soit récemment introduits.


##### Conclusion business :

- Headphones = produit clé
